```python
happiness= pd.read_csv('data/happiness_survey_data.csv')
happiness.head()
```

```python
(happiness[happiness.country_name
    .isin(['Canada', 'Mexico', 'United States'])]
    .iloc[:, :3]
    .pivot(index='year', columns='country_name', values='happiness_score').plot.line())
```
